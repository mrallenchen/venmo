### 1. Get a list of names
- The list of names were generated by looking at the most common American names as published by the U.S. Social Security Administration for the last 6 decades. (This was done manually and saved in the csv file)
- Pages: Count of pages to pull. Since Venmo provides search results broken up by pages, the page count in the csv indicates how many pages worth of usernames we want for each search.
    - The number of pages was determined by weighting the name and how common it is in the U.S.

```python
names = pd.read_csv("Names.csv",sep="\t")
```

```python
names[0:5]
```

### 2. Generate list of usernames by searching names through Venmo API

```python
user_list = []
for i in range(len(names)):
    for j in range(1,(names.Pages[i])):
        print(names.Names[i],j)
        user_list.extend(venmo.user.search_for_users(query=names.Names[i],page=j))
```

```python
# save the list to be used for later
with open('data/user_list.pickle', 'wb') as f:
    pickle.dump(user_list, f)
```

When this was pulled on 8/9/2020, this generated 29,850 users!

## 3. Pull Transactions for each User

```python
#note that some users will generate an error. This may happen if the user is no longer considered active.
#In any case, the code will just pass over them and note it in an error_list for reference if needed in the future.

counter = 0
error_list = []
for user in user_list[44:]:
    counter += 1
    try:
        transactions.extend(venmo.user.get_user_transactions(user_id=user.id))
    except:
        error_list.append(user)
        print(f'error with {user.username} {user.id}')
    if counter % 10 == 0:
        print(user.username, counter)
```
