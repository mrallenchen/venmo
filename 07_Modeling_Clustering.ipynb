### Identify empty clusters, where there is a centroid but no item attached.

```python
cluster_labels = list(pd.Series(labels).unique())
```

```python
clusters_empty_check = [x not in cluster_labels for x in list(range(NUM_CLUSTERS))]
```

```python
empty_clusters = []
for i,x in enumerate(clusters_empty_check):
    if x:
        empty_clusters.append(i)
```

```python
empty_clusters_count = pd.Series(np.zeros(len(empty_clusters)),index=empty_clusters)
```

```python
empty_clusters_count
```

```python
note_count = pd.concat([cluster_counts.iloc[:,0],(empty_clusters_count)])
```

### Analysis of Clusters

```python
#note: do not user np.argsort because that will give arguments that are not associated with the indexes
display_order = list(note_count.sort_values(ascending=False).index)
```

```python
cluster_content = full_df.groupby('cluster').sum().transpose()
```

```python
cluster_content
```

```python
ordered_cluster_content = []
for i in range(NUM_CLUSTERS):
    #note that need to use a try and except here, because kmeans sometimes creates centroids that have no documents
    try:
        ordered_cluster_content.append(
        list(zip(list(
        cluster_content[i].sort_values(ascending=False).index),
        cluster_content[i].sort_values(ascending=False))
        ))
    except:
        ordered_cluster_content.append(np.zeros(num_words)
        )
```

```python
ordered_cluster_words = []
for i in range(NUM_CLUSTERS):
    #note that need to use a try and except here, because kmeans sometimes creates centroids that have no documents
    try:
        ordered_cluster_words.append(
        cluster_content[i].sort_values(ascending=False).index)
    except:
        ordered_cluster_words.append(np.zeros(num_words)
        )
```

```python
framed_cluster_words = pd.DataFrame(ordered_cluster_words)
```

```python
framed_cluster_words.iloc[display_order,]
```

```python
ordered_cluster_counts = []
for i in range(NUM_CLUSTERS):
    #note that need to use a try and except here, because kmeans sometimes creates centroids that have no documents
    try:
        ordered_cluster_counts.extend(
            cluster_content[i].sort_values(ascending=False))
    except:
        ordered_cluster_counts.extend(np.zeros(num_words))
framed_cluster_counts = np.array(ordered_cluster_counts).reshape(len(note_count),len(columns))
```
